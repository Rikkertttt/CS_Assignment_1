```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("istherecorrelation.csv", delimiter=";")
df["WO"] = df["WO [x1000]"].str.replace(',', '.').astype(float)
df["Consumption (hl)"] = df["NL Beer consumption [x1000 hectoliter]"] * 1000
```

```python
fig, ax = plt.subplots()
corr = df["WO"].corr(df["Consumption (hl)"])
sns.regplot(data=df, x="WO", y="Consumption (hl)", scatter=True, ax=ax, color='red')
ax.text(0.05, 0.95, f"Correlation: {corr:.2f}", transform=ax.transAxes, fontsize=12, verticalalignment='top')
ax.set_xlabel("# of WO students (x1000)")
ax.set_ylabel("Total NL beer consumption (hectolitres)")
plt.show()
```

```python
fig, ax = plt.subplots()
corr = df["Year"].corr(df["Consumption (hl)"])
sns.regplot(data=df, x="Year", y="Consumption (hl)", scatter=True, ax=ax, color='red')
ax.text(0.05, 0.95, f"Correlation: {corr:.2f}", transform=ax.transAxes, fontsize=12, verticalalignment='top')
ax.set_xlabel("Year")
ax.set_ylabel("Total NL beer consumption (hectolitres)")
plt.show()
```
